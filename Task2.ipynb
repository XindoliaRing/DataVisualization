```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib as mpl
from scipy.stats import norm

# 中文乱码的处理
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']# 设置微软雅黑字体
plt.rcParams['axes.unicode_minus'] = False # 避免坐标轴不能正常的显示负号

fig, axes = plt.subplots(2, 3, figsize=(15, 10)) 
ax1, ax2, ax3, ax4, ax5, ax6 = axes.flatten()

# 绘制第 1 个子图：条形图
ax1 = plt.subplot(231)  
plt.bar([1, 2, 3], [0.6, 0.8, 0.2], color=['plum','paleturquoise','moccasin'])
ax1.set_title('三种不同颜色的条形图')

# 绘制第 2 个子图：线图
ax2 = plt.subplot(232)  
x = np.linspace(0, 10, 30)
plt.plot(x, np.sin(x), color='mediumpurple', linestyle='-', marker='.', label='sin(x)')  
plt.plot(x, np.cos(x), color='cornflowerblue', linestyle='--', marker='*', label='cos(x)') 
plt.legend(loc='upper right')
ax2.set_title('sin(x)和cos(x)的图像')

# 绘制第 3 个子图：水平条形图
ax3 = plt.subplot(233)  
ylabel = ['A','B','C','D','E','F','G','H']
xnum = [3,1,4,5,8,9,7,2]
# for x,y in enumerate(xnum):
#    plt.text(y+0.5, x, '%s'%y, va='center')
plt.barh(ylabel, xnum)
plt.xlim([0, 10])
ax3.set_title('水平条形图')

# 绘制第 4 个子图：气泡图
ax4 = plt.subplot(234)
colors = np.random.randn(100)
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, 
            #大小
            s = np.power(10*x + 20*y, 2),
            #颜色
            c = colors,
            #标注
            marker = 'o',
            #指定某个colormap值
            cmap=mpl.cm.RdYlBu,
            #透明度
            alpha = 0.9)
ax4.set_title('气泡图')

# 绘制第 5 个子图：棉棒图
ax5 = plt.subplot(235)
x5 = np.linspace(0.5, 2*np.pi, 20)
y5 = np.random.randn(20)
plt.stem(x5, y5, linefmt="-.", markerfmt="o", basefmt="-")
ax5.set_title('棉棒图')

# 绘制第 6 个子图：双层环形图
ax6 = plt.subplot(236)
weight1 = [40, 15, 20, 10, 15]
weight2 = [30, 25, 15, 20, 10]
weight3=[1]
color=["#e41a1c","#377eb8","#4daf4a","#984ea3","#ff7f00"]
labels = ["Flour","Sugar","Cream","Strawberry","Nuts"]
ax6.pie(weight1, radius=1.2, autopct='%1.1f%%', pctdistance=0.9, colors=color, wedgeprops = {'linewidth': 1, 'edgecolor':'cornflowerblue'})
ax6.pie(weight2, radius=1, autopct='%1.1f%%', pctdistance=0.75, colors=color, wedgeprops = {'linewidth': 1, 'edgecolor':'cornflowerblue'})
ax6.pie(weight3, radius=0.5, colors='w')
ax6.set(aspect="equal", title='双层环形图')

plt.show()
```
