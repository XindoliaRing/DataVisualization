### 第一题：画出直线 y = x-1, 线型为虚线，线宽为1，纵坐标范围（-2，1），横坐标范围（-1，2），横纵坐标在（0，0）坐标点相交。横坐标的 [-1,-0.5,1] 分别对应 [bad, normal, good]。

```python
x3 = np.linspace(-1, 2, 50)
y3 = x3 - 1

plt.figure()
plt.plot(x3, y3,linewidth=1.0, linestyle='--', label='linear line')
plt.xlim((-1, 2))
plt.ylim((-2, 1))

plt.xlabel('X')
plt.ylabel('Y')

NewTicks = np.linspace(-2, 1, 4)
plt.yticks(NewTicks)
plt.xticks([-1, -0.5, 1], [r'$bad$', r'$normal$', r'$good$'])

ax3 = plt.gca()
ax3.spines['right'].set_color('none')
ax3.spines['top'].set_color('none')

ax3.xaxis.set_ticks_position('bottom')
ax3.spines['bottom'].set_position(('data', 0))
ax3.yaxis.set_ticks_position('left')
ax3.spines['left'].set_position(('data',0))

plt.show()
```

### 第二题：假设大家在30岁的时候，根据自己的实际情况，统计出来了你和同桌从11岁到30岁每年交的朋友的数量如列表a和b,请绘制出该数据的折线图,以便分析自己和同桌每年交朋友的数量走势

a = [1,0,1,1,2,4,3,2,3,4,4,5,6,5,4,3,3,1,1,1]

b = [1,0,3,1,2,2,2,3,1,1,1,1,1,2,1,1,2,3,2,2]

y轴表示个数

x轴表示岁数,比如11岁,12岁

```python
plt.figure(figsize = (20, 10))
a = [1, 0, 1, 1, 2, 4, 3, 2, 3, 4, 4, 5, 6, 5, 4, 3, 3, 1, 1, 1]
b = [1, 0, 3, 1, 2, 2, 2, 3, 1, 1, 1, 1, 1, 2, 1, 1, 2, 3, 2, 2]
x = list(range(11, 31))
x_l = [str(i) + "岁" for i in x]
plt.plot(x, a, label = "自己")
plt.plot(x, b, label = "同桌")
co_a = zip(x, a)
co_b = zip(x, b)
plt.xticks(x, x_l, rotation = 45)
plt.legend()
plt.grid(alpha = 0)
plt.xlabel("岁数")
plt.ylabel("个数")
plt.title("自己和同桌每年交朋友的数量走势情况图")
plt.show()
```
